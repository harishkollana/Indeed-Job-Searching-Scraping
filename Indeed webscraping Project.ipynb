```python
#import libraries to webscrape and save data as csv
import requests
from bs4 import BeautifulSoup
import csv

#create a list of urls to scrape
urls = ['https://www.indeed.com/jobs?q=data+scientist&l=New+York%2C+NY','https://www.indeed.com/jobs?q=data+analyst+entry+level&l=New+York%2C+NY']

#create a list to store the data
data = []

#loop through the urls
for url in urls:
    #make a request to the url
    response = requests.get(url)
    #parse the html
    soup = BeautifulSoup(response.text, 'html.parser')
    #find the job listings
    job_listings = soup.find_all('div', class_='job_seen_beacon')
    #loop through the job listings
    for job_listing in job_listings:
        #create a dictionary to store the data
        job_data = {}
        #find the title
        title = job_listing.find('h2', class_='jobTitle')
        #add the title to the dictionary
        job_data['title'] = title.text
        #find the company
        company = job_listing.find('span', class_='companyName')
        #add the company to the dictionary
        job_data['company'] = company.text
        #add the dictionary to the data list
        data.append(job_data)

#create a csv file to store the data
with open('indeed_data.csv', 'w') as csvfile:
    #create the csv writer object
    csvwriter = csv.writer(csvfile)
    #write the header
    csvwriter.writerow(['title', 'company'])
    #write the data
    for job in data:
        csvwriter.writerow([job['title'], job['company']])

#print the data
print(data)
```
